```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nengo
import scipy
import pandas as pd

palette = sns.color_palette('tab10')
sns.set_palette(palette)
sns.set(context='paper', style='whitegrid', font="cmr10", font_scale=1.0)
plt.rcParams['axes.formatter.use_mathtext'] = True
```

```python
rl_values = pd.read_pickle("data/rl_values.pkl")
```

```python
monkey = 'V'
session = 0
block = 19
data = rl_values.query("monkey==@monkey & session==@session & block==@block")

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(14,4), sharex=True, sharey=True)
sns.lineplot(data, x='trial', y='V_StimL', ax=axes[0], color=palette[0], label='V_StimL')
sns.lineplot(data, x='trial', y='V_StimR', ax=axes[0], color=palette[1], label='V_StimR')
sns.lineplot(data, x='trial', y='V_Act1', ax=axes[0], color=palette[2], label='V_Act1')
sns.lineplot(data, x='trial', y='V_Act2', ax=axes[0], color=palette[3], label='V_Act2')
sns.lineplot(data, x='trial', y='omega', ax=axes[1], color=palette[0])
sns.lineplot(data, x='trial', y='DV_left', ax=axes[2], color=palette[0], label='DV_left')
sns.lineplot(data, x='trial', y='DV_right', ax=axes[2], color=palette[1], label='DV_right')
sns.lineplot(data, x='trial', y='reward', ax=axes[3], color=palette[0])
axes[0].set(title='value', xlabel='RL variable')
axes[1].set(title='omega')
axes[2].set(title='DV')
axes[3].set(title='reward')
fig.tight_layout()
```
